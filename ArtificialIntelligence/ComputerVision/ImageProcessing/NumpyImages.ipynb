# Numpy

```python
import numpy as np
```

```python
# Type
a = np.array([2,3,4])
print('Type :', type(a))

# range of Numpy -> arrange()
print(np.arange(0,10,2), '\n')

# (row, column)
print(np.ones(shape=(3,4)), '\n')
print(np.zeros(shape=(2,4)), '\n')

mat = np.arange(0,100).reshape(10,10)
print('10 * 10 Matrix:\n', mat)


# Indexing
print('\nIndex (5,5) :', mat[5,5])
print('\nThird Column :', mat[:,2].reshape(10,1))
print('\nThird row :', mat[2,:])

# Slicing exclusive index
print('\nSliced 3*5 Matrix:\n', mat[0:3,0:5])

# Modification
mat_copy = mat.copy()
mat_copy[0:5,0:10] = 0
print('\nModified Matrix:\n', mat_copy)
```

```python
# Information About the array
def infoArr(arr):
    print('\nType :' ,type(arr))
    print('\nShape :' ,arr.shape)
    print('\nDimensions :' ,arr.ndim)
    print('\nDtype:' ,arr.dtype.name)
    print('\nMax Value :' ,arr.max())
    print('\nMin Value :' ,arr.min())
    # print('\nLocation min:' ,arr.argmax())
    print('\nMemory Loc:' ,arr.data)
    return None
```

### 3-Channel RGB Image/ Info

```python
pic = Image.open('image.png')
print(type(pic), '\n\t|\n After Conversion to Numpy Array : \n\t|\n\tv')
pic_np = np.asarray(pic)
pic_rgb = pic_np.copy()

print('Image(Numpy) : \n\n', plt.imshow(pic_np))

# Grayscale RGB
def plot_rgb(pic):
    fig = plt.figure()
    for  i in range(3):
        a = fig.add_subplot(1,3,i+1)
        implot = plt.imshow(pic_rgb[:,:,i], cmap='gray')
        if i == 0:
            a.set_title('Red')
            
        elif i == 1:
            a.set_title('Green')
        else:
            a.set_title('Blue')
        
plot_rgb(pic_rgb)
```

### Accessing RGB Channels

```python
# Importing image

pic = Image.open('./ImgResources/profile.png')
pic_rgb = np.asarray(pic).copy()

infoArr(pic_rgb)
print('Image(Numpy) : \n\n', plt.imshow(pic_rgb))

pic_rgb[:,:,0] = 0 # Switch OFF Red channel
pic_rgb[:,:,2] = 0 # Switch OFF Blue channel 
plot_rgb(pic_rgb)
```
