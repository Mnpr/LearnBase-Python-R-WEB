# Image Processing

## Opening Images

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import cv2

img = cv2.imread('./ImgResources/owl.jpeg')

# plt.imshow(img)

# Matplotlib -> RGB
# OpenCV -> BGR

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB )
print(img_rgb.shape)

plt.imshow(img_rgb)
```

```python
# Grayscale single channel Images read

img_gray = cv2.imread('./ImgResources/owl.jpeg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray, cmap='gray')

print(img_gray.shape)
```

```python
# Resize the imgaes

# resized_img = cv2.resize(img_rgb, (200, 300)) # (y,x) size

w_ratio = 0.6 # 60% of original width
h_ratio = 0.8 # 80% of original height

resized_img= cv2.resize(img_rgb,(0,0), img_rgb, w_ratio, h_ratio )
plt.imshow(resized_img)
```

```python
# Flip Imgaes

flipped_img = cv2.flip(img_rgb, -1) # 0 - Horizontal, 1 - Vertical Flip (BOTH Flips : -1)
plt.imshow(flipped_img)
```

```python
# Write Images
cv2.imwrite('./ImgResources/flipped_owl.jpg', flipped_img)
```

```python
# Canvas Config Matplotlib

fig = plt.figure(figsize = (2,2))
ax = fig.add_subplot(111)
ax.imshow(img_rgb)
```

### HSL and HSV colormappings

```python
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv.imread('./ImgResources/profile.png')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

plt.imshow(img)
```

```python
# HSV
img = cv.cvtColor(img, cv.COLOR_RGB2HLS_FULL)
plt.imshow(img)
```

```python
# HSL
img = cv.cvtColor(img, cv.COLOR_RGB2HLS_FULL)
plt.imshow(img)
```
