# Blurring and Smoothing

- Often Blurring and Smoothing is combined with Edge Detection
- Gamma Correction method. (Gamma Value - brighter/darker)
- Kernel Based Filters

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load_img(img_loc):
    img = cv.imread(img_loc).astype(np.float32) / 255
    img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
    return img

def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
    
monkey = load_img('./ImgResources/monkey.png')
display_img(monkey)
```

```python
# Gamma Correction

gamma = 0.5 # (less than 1 = bright | greater than 1 = darker)
result = np.power(monkey, gamma)
display_img(result)
```

```python
monkey = load_img('./ImgResources/monkey.png')

font = cv.FONT_HERSHEY_COMPLEX
cv.putText(monkey, text='ZEN',org=(10,590), fontFace=font, fontScale=9, color=(255,0,0), thickness=4)

display_img(monkey)
```

```python
# Blurring : Low Pass filter | 2D convolution

kernel = np.ones(shape=(5,5), dtype=np.float32)/ 25
output_img = cv.filter2D(monkey, -1, kernel) # desired depth -1 => i/p depth = o/p depth
display_img(output_img)
```

```python
# Blur/smooth using averaging
blurred_img = cv.blur(monkey, ksize=(10,10))
display_img(blurred_img)
```

```python
# Gaussian Blurring
blurred_img = cv.GaussianBlur(monkey, (5,5), 10) # S.D. : 10
display_img(blurred_img)
```

```python
# Medial Blur : Reducing Img Noise
blurred_img = cv.medianBlur(monkey, 5)
display_img(blurred_img)
```

```python
# Bi-lateral Filtering
blurred_img = cv.bilateralFilter(monkey,9,75,75)
display_img(blurred_img)
```
