# Gradients

- Extension of Morphological operators.
- An image gradient is a directional change in the ingensity or color of an image.
- Gradients can be calculated in specific direction.
    - Normalized x-gradient from SF-Operator - Vertical Edge detection
    - Normalized y-gradient from SF-Operator - Horizontal Edge detection
    
**Exploration of Sobel-Feldman Operators**

- The operator uses two $3*3$ kernels which are convolved with the original image to calculate approximations of the derivatives(rate of change in intensity). ( One for horizontal edge, and another fro vertical)

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def display_img(img):
    fig = plt.figure(figsize=(8,6))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
img = cv.imread('./ImgResources/peace.jpeg',0) # Red in Grayscale
display_img(img)
```

```python
# Vertical Edges

# depth 64F Precision, dx direction =1, dy = 0, kernel size
sobel_x = cv.Sobel(img,cv.CV_64F, 1,0, ksize=5)
display_img(sobel_x)
```

```python
# Horizontal Edges

# depth 64F Precision, dx direction =1, dy = 0, kernel size
sobel_y = cv.Sobel(img,cv.CV_64F, 0,1, ksize=5)
display_img(sobel_y)
```

```python
# Laplacian Gradient

laplacian = cv.Laplacian(img, cv.CV_64F)
display_img(laplacian)
```

```python
# Blending Vertical and Horizontal Edges

blended = cv.addWeighted(src1=sobel_x, alpha=0.5, src2=sobel_y, beta=0.5, gamma=0)
display_img(blended)
```

```python
# Thresholding

ret, th1 = cv.threshold(img, 100, 255, cv.THRESH_BINARY)
display_img(th1)
```
