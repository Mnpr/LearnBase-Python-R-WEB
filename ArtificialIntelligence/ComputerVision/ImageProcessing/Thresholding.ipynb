# Thresholding

```python
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline

img = cv.imread('./ImgResources/profile.png',0)

def show_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
    
show_pic(img)
```

```python
ret,th1 = cv.threshold(img,128,255, cv.THRESH_BINARY)
show_pic(th1)
```

```python
# Adaptive thresholding
th2 = cv.adaptiveThreshold(img, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY,11,8)
show_pic(th2) 
```

```python
# Blend different thresholds
blended = cv.addWeighted(src1=th1, alpha=0.6, src2=th2, beta=0.4, gamma=0)
show_pic(blended)
print('Shape: ', blended.shape)
```
